```python
# @title Update favorite_program for selected student in JSON
name = "Asaad"  # @param ["Asaad","Qasem","Marwan","Joul"]
change = "dffdfdfjhhk "    # @param {"type":"string"}

import json
from pathlib import Path

FILE_PATH = Path("/content/drive/My Drive/students.json")

# Try mounting Drive if file isn't found
if not FILE_PATH.exists():
    try:
        from google.colab import drive
        drive.mount('/content/drive')
    except Exception as e:
        print("Couldn't auto-mount Drive. If needed, mount manually: from google.colab import drive; drive.mount('/content/drive')")

if not FILE_PATH.exists():
    raise FileNotFoundError("⚠️ students.json file not found at /content/drive/My Drive/Colab Notebooks/students.json")

# Read JSON list
with open(FILE_PATH, "r", encoding="utf-8") as f:
    students = json.load(f)

found_idx = -1
before_obj = None

for i, s in enumerate(students):
    # assuming the JSON has keys like "first_name" and "last_name"
    first_name = s.get("first_name", "").strip()
    if first_name.lower() == name.lower():
        found_idx = i
        before_obj = s.copy()
        break

if found_idx == -1:
    print("Student not found in JSON.")
else:
    change_clean = change.strip()
    if change_clean:
        # set / override favorite_program field
        students[found_idx]["favorite_program"] = change_clean

        # save back to JSON
        with open(FILE_PATH, "w", encoding="utf-8") as f:
            json.dump(students, f, ensure_ascii=False, indent=2)

        print("✅ Student Found and Updated")
        print("Before:\n", before_obj)
        print("\nAfter:\n", students[found_idx])
    else:
        print("✅ Student Found (no change provided, JSON not modified)")
        print("Current object:\n", students[found_idx])
```
