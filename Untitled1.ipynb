```python
# Dynamic widget version – works with any names in students.json

from google.colab import drive
drive.mount('/content/drive')

import json
from pathlib import Path
import ipywidgets as widgets
from IPython.display import display

# Path to your JSON file
FILE_PATH = Path("/content/drive/My Drive/students2.json")

if not FILE_PATH.exists():
    raise FileNotFoundError(f"students.json not found at: {FILE_PATH}")

# Load students from JSON (list of objects)
with open(FILE_PATH, "r", encoding="utf-8") as f:
    students = json.load(f)

# Build dynamic name list + index mapping
name_to_idx = {}
options = []

for i, s in enumerate(students):
    # Try first_name/last_name, fall back to 'name' if needed
    first = s.get("first_name") or s.get("name") or f"Student{i+1}"
    last = s.get("last_name", "")
    full_name = f"{first} {last}".strip()
    name_to_idx[full_name] = i
    options.append(full_name)

# --- Widgets (instead of @param) ---
name_dropdown = widgets.Dropdown(
    options=options,
    description='Student:',
    layout=widgets.Layout(width='250px')
)

change_box = widgets.Text(
    description='Favorite:',
    placeholder='Enter favorite program',
    layout=widgets.Layout(width='300px')
)

update_button = widgets.Button(
    description='Update',
    button_style='success',
    layout=widgets.Layout(width='120px')
)

output = widgets.Output()

# --- Update logic ---
def on_update_clicked(b):
    with output:
        output.clear_output()

        selected_name = name_dropdown.value
        change = change_box.value.strip()

        if not change:
            print("⚠️ No text entered for favorite program.")
            return

        idx = name_to_idx[selected_name]
        before_obj = students[idx].copy()

        # Update / create favorite_program field
        students[idx]["favorite_program"] = change

        # Save back to JSON
        with open(FILE_PATH, "w", encoding="utf-8") as f:
            json.dump(students, f, ensure_ascii=False, indent=2)

        print("✅ Updated JSON for:", selected_name)
        print("\nBefore:\n", before_obj)
        print("\nAfter:\n", students[idx])

update_button.on_click(on_update_clicked)

# Show the UI
display(name_dropdown, change_box, update_button, output)
```
